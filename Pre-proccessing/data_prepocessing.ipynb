```python
import pandas as pd

# Read the data after we imputed the age column with the mean
df = pd.read_csv('Data/colon-dataset.csv', encoding_errors='ignore')

df
```

```python
# Drop the columns that can cause imbalanced data
df.drop(['trail', 'CD28','pd1duz','pdl1poly','CD40','cd27snp'], axis=1, inplace=True)
df
```

```python
#checking for null values
df.isnull().sum()
```

```python
#rename the column to make more sense
df.replace('kontrol', 'healthy', inplace=True)
df
```

```python
#rename the column to make more sense
df.replace('kolon ca', 'diagnosed', inplace=True)

df
```

```python
#hot encoding the categorical data
columns_to_encode = ['p16540', 'p16580', 'mdm2', 'GAL3', 'TIM1']

df_encoded = pd.get_dummies(df, columns=columns_to_encode)

df_encoded
```

```python
#rename field 
df_encoded.rename(columns={'kolongrup':'Class'}, inplace=True)

df_encoded
```

```python
#double checking for null values
df_encoded.isnull().sum()
```
